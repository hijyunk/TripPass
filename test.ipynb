```python
import os
from serpapi import GoogleSearch

params = {
    "engine": "google_maps",
    "q": "attractions in New York",
    "type": "search",
    "google_domain": "google.com",
    "hl": "en",
    "api_key": os.environ.get('SERPAPI_API_KEY')
}

search = GoogleSearch(params)
results = search.get_dict()
```

```python
import json
print(json.dumps(results, ensure_ascii=False, indent=3))
```
